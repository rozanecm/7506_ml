```python
import pandas as pd
```

```python
df = pd.read_csv('data/events.csv', dtype={'sku':object})
```

```python
df.head()
```

```python
# some date processing
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['weekday'] = df['timestamp'].dt.day_name()
df['hour'] = df['timestamp'].dt.hour
df['year_month_day'] = df['timestamp'].map(lambda x: str(x.year)+"/"+str(x.month)+"/"+str(x.day))
df['year_month_day'] = pd.to_datetime(df['year_month_day'])
```

```python
def to_lower(obj):
    if type(obj) is str:
        return obj.lower()
    else:
        return obj
    

df.model.apply(lambda x: to_lower(x))
```
